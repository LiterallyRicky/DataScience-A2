**Task5**:

Create and run a stright line linear regression model on the howell data set with height predicting weight. Visualize your model and its posterior.

```python
#lb stands for linear bike
with pm.Model() as model_howell1:

    #priors for the linear part of our model
    α = pm.Normal("α", mu=0, sigma=100)
    β = pm.Normal("β", mu=0, sigma=10)

    #this is the linear part of our model
    μ = pm.Deterministic("μ", α + β * howell.height)

    #prior for the standard deviation of our likelihood
    #Cauchy is a T dist with nu = 1
    σ = pm.HalfCauchy("σ", 10)

    #likelihood
    y_pred = pm.Normal("y_pred", mu=μ, sigma=σ, observed=howell.weight)

    #inference data object
    idata_howell1 = pm.sample()
```

```python
az.plot_posterior(idata_howell1, var_names=["~μ"], figsize=(12, 3))
```

```python
_, ax = plt.subplots()

ax.plot(howell.height, howell.weight, "C0.", alpha=0.1)

posterior_howell = az.extract(idata_howell1)

μ_m = posterior_howell["μ"].mean("sample").values
σ_m = posterior_howell["σ"].mean("sample").values

ax.plot(howell.height, μ_m, c="k")
ax.fill_between(howell.height, μ_m + 1 * σ_m, μ_m - 1 * σ_m, alpha=0.6, color="C1")
ax.fill_between(howell.height, μ_m + 2 * σ_m, μ_m - 2 * σ_m, alpha=0.4, color="C1")

ax.set_xlabel("height")
ax.set_ylabel("weight")
```

**Task6**:

Run a posterior predictive check for model_howell, plot the posterior predictive distribution, and comment on the fit of the model you made in Task5.

```python
pm.sample_posterior_predictive(idata_howell1, model = model_howell1, extend_inferencedata=True)
```

```python
az.plot_ppc(idata_howell1, num_pp_samples=200, colors=["C1", "C0", "C1"])
```

So this PPC is actually not too bad but can definitely be better. Some things that I like about this is the way the shape is generally the same towards the peak. The right tail of the distribution is modeled nice as well. However, the observed seems to have a smaller peak towards the left of the distribution. To improve this PPC, our next model should match BOTH peaks rather than just one. :D

**Task7**:

Create and visualize the posterior of a generalized linear model that improves the on the fit of the model you made in Task5.

```python
with pm.Model() as improved_model:
   #priors for the linear part of our model
    α = pm.Normal("α", mu=0, sigma=10)
    β = pm.Normal("β", mu=0, sigma=10)

    #this is the linear part of our model
    μ = pm.Deterministic("μ", α + β * howell.height**3)

    #prior for the standard deviation of our likelihood
    #Cauchy is a T dist with nu = 1
    σ = pm.HalfCauchy("σ", 10)

    #likelihood
    y_pred = pm.Normal("y_pred", mu=μ, sigma=σ, observed=howell.weight)

    #inference data object
    idata_improved = pm.sample()
```

```python
az.plot_posterior(idata_improved, var_names=["~μ"], figsize=(12, 3))
```

```python
_, ax = plt.subplots()

ax.plot(howell.height, howell.weight, "C0.", alpha=0.1)

posterior_howell = az.extract(idata_improved)

μ_m = posterior_howell["μ"].mean("sample").values
σ_m = posterior_howell["σ"].mean("sample").values

ax.plot(howell.height, μ_m, c="k")
ax.fill_between(howell.height, μ_m + 1 * σ_m, μ_m - 1 * σ_m, alpha=0.6, color="C1")
ax.fill_between(howell.height, μ_m + 2 * σ_m, μ_m - 2 * σ_m, alpha=0.4, color="C1")

ax.set_xlabel("height")
ax.set_ylabel("weight")
```

**Task8**:

Run a posterior predictive check, and visualize the posterior predictive distribution, for the model you made in Task7.

```python
pm.sample_posterior_predictive(idata_improved, model = improved_model, extend_inferencedata=True)
```
